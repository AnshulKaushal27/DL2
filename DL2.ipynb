```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 100 random values between -10000 and 10000
x = np.random.uniform(-10000, 10000, 100)

# Sort values for better plotting
x = np.sort(x)

# Activation Functions
def sigmoid(x):
    return 1/(1 + np.exp(-x))

def sigmoid_derivative(x):
    s = sigmoid(x)
    return s * (1 - s)

def tanh(x):
    return np.tanh(x)

def tanh_derivative(x):
    return 1 - np.tanh(x)**2

def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return np.where(x > 0, 1, 0)

def leaky_relu(x):
    return np.where(x > 0, x, 0.01*x)

def leaky_relu_derivative(x):
    return np.where(x > 0, 1, 0.01)

def softplus(x):
    return np.log(1 + np.exp(x))

def softplus_derivative(x):
    return sigmoid(x)

def elu(x, alpha=1):
    return np.where(x > 0, x, alpha*(np.exp(x)-1))

def elu_derivative(x, alpha=1):
    return np.where(x > 0, 1, alpha*np.exp(x))


# Plot Activation Functions
plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
plt.plot(x, sigmoid(x), label="Sigmoid")
plt.plot(x, tanh(x), label="Tanh")
plt.plot(x, relu(x), label="ReLU")
plt.plot(x, leaky_relu(x), label="Leaky ReLU")
plt.plot(x, softplus(x), label="Softplus")
plt.plot(x, elu(x), label="ELU")

plt.title("Activation Functions")
plt.legend()
plt.grid()

# Plot Derivatives
plt.subplot(1,2,2)
plt.plot(x, sigmoid_derivative(x), label="Sigmoid Derivative")
plt.plot(x, tanh_derivative(x), label="Tanh Derivative")
plt.plot(x, relu_derivative(x), label="ReLU Derivative")
plt.plot(x, leaky_relu_derivative(x), label="Leaky ReLU Derivative")
plt.plot(x, softplus_derivative(x), label="Softplus Derivative")
plt.plot(x, elu_derivative(x), label="ELU Derivative")

plt.title("Derivatives of Activation Functions")
plt.legend()
plt.grid()

plt.show()
```
